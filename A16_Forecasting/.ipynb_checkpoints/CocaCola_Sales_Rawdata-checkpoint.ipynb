## Holt-Winters Methods(i.e. Exponential Smoothing Techniques)

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.metrics import mean_squared_error
```

### Simple Exponential Smoothing

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

alpha = 0.5
```

```python
# Decompose the data frame to get the trend, seasonality and noise
decompose_result = seasonal_decompose(data1['Sales'],period =4 ,model='add')
fig = decompose_result.plot()
fig.set_size_inches((16, 9))
plt.show()
```

```python
model_ses12=SimpleExpSmoothing(data1['Sales'])
fitted_model= model_ses12.fit(smoothing_level=alpha,optimized=False)
data1['SES12'] = fitted_model.fittedvalues.shift(-1)
data1.head()
```

```python
ses12_mse = mean_squared_error(data1['Sales'][:41],data1['SES12'][:41])
ses12_rmse = np.sqrt(ses12_mse)
ses12_mse,ses12_rmse
```

### Double Exponential Smoothing

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

mod_add12 = ExponentialSmoothing(data1['Sales'], trend='add')
fitted_model= mod_add12.fit()
data1['DESadd12'] = fitted_model.fittedvalues.shift(-1)
data1
```

```python
data1[['Sales','SES12','DESadd12']].plot(figsize=(12,6)).autoscale(axis='x',tight=True);
```

```python
data1[['Sales','SES12','DESadd12']].iloc[:24].plot(figsize=(12,6)).autoscale(axis='x',tight=True);
```

```python
data1['DESmul12'] = ExponentialSmoothing(data1['Sales'], trend='mul').fit().fittedvalues.shift(-1)
data1.head()
```

```python
data1[['Sales','DESadd12','DESmul12']].iloc[:24].plot(figsize=(12,6)).autoscale(axis='x',tight=True);
```

```python
desadd12_mse = mean_squared_error(data1['Sales'][:41],data1['DESadd12'][:41])
desadd12_rmse = np.sqrt(desadd12_mse)
desadd12_mse,desadd12_rmse
```

```python
desmul12_mse = mean_squared_error(data1['Sales'][:41],data1['DESmul12'][:41])
desmul12_rmse = np.sqrt(desmul12_mse)
desmul12_mse,desmul12_rmse
```

## AR, MA, ARMA Models

### The ADF-Test(i.e. Test for Stationarity)

```python
data2 = data.copy()
data2.head()
```

```python
train_data = data2.iloc[:32] 
test_data = data2.iloc[32:]
train_data
```

```python
from statsmodels.tsa.stattools import adfuller
def adf_test(series):    
    result = adfuller(series.dropna())  
    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)
    #for key,val in result[4].items():
        #out[f'critical value ({key})']=val
    if result[1] <= 0.05:
        print("Reject the null hypothesis")
        print("Data is stationary")
    else:
        print("Fail to reject the null hypothesis")
        print("Data is non-stationary")
```

```python
adf_test(train_data['Sales'])
```

+ as the p-value of ADF-Test is more than 0.05(i.e. >0.05) and the Data is not Stationary, therefore we cannot use AR, MA, ARMA models for Forecasting this data. So we use ARIMA Model

## ARIMA Model

```python
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
data2 = data.copy()
data2.head()
```

```python
result = seasonal_decompose(data2['Sales'], model='mul')
fig = result.plot()
fig.set_size_inches((16, 9))
plt.show()
```

### order of differencing (d) in ARIMA model

#### Do first order differencing and do ADF-Test 

```python
from statsmodels.tsa.statespace.tools import diff
data2['diff1'] = diff(data2['Sales'],k_diff=1)
adf_test(data2['diff1'])
```

```python
data2['diff2'] = diff(data2['diff1'],k_diff=1)
adf_test(data2['diff2'])
```

```python
data2
```

```python
# Original Series
fig, axes = plt.subplots(3, 2,figsize=(15,15))

axes[0, 0].plot(data2.Sales); axes[0, 0].set_title('Original Series')
plot_acf(data2.Sales, ax=axes[0, 1])

# 1st Differencing
axes[1, 0].plot(data2.Sales.diff()); axes[1, 0].set_title('1st Order Differencing')
plot_acf(data2.Sales.diff().dropna(), ax=axes[1, 1])

# 2nd Differencing
axes[2, 0].plot(data2.Sales.diff().diff()); axes[2, 0].set_title('2nd Order Differencing')
plot_acf(data2.Sales.diff().diff().dropna(), ax=axes[2, 1])

plt.subplots_adjust(left=0.1,bottom=0.2, right=0.9, top=1, wspace=0.5, hspace=0.5)
plt.show()
```

+ We use 2nd order Differencing for our Dataset because  on looking at the autocorrelation plot for the 2nd differencing is Stationary.

```python
data3 = data.copy()
```

+ I take p as 1 or 2, as we can see from PACF-Plot

### order of the MA term (q)

```python
# ACF plot of 1st and 2nd differenced series
fig, axes = plt.subplots(2, 2, figsize=(15,4))
axes[0,0].plot(data3.Sales.diff()); axes[0,0].set_title('1st Differencing')
axes[0,1].set(ylim=(-2,2))
plot_acf(data3.Sales.diff().dropna(), ax=axes[0,1])

axes[1,0].plot(data3.Sales.diff().diff()); axes[1,0].set_title('2st Differencing')
axes[1,1].set(ylim=(-2,2))
plot_acf(data3.Sales.diff().dropna(), ax=axes[1,1])

plt.subplots_adjust(left=0.1,bottom=0.2, right=0.9, top=1, wspace=0.5, hspace=0.5)
plt.show()
```

### Final ARIMA Model

```python
# 2,2,2 ARIMA Model
arima_model = ARIMA(train_data, order=(2,2,2))
model_fit = arima_model.fit()
```

```python
test_data.shape
```

```python
# Forecast
fc, se, conf = model_fit.forecast(8, alpha=0.05)
test_predictions = pd.Series(fc, index=test_data.index)
test_predictions
```

```python
# Plot
plt.figure(figsize=(12,5), dpi=100)
plt.plot(train_data, label='training')
plt.plot(test_data, label='test')
plt.plot(test_predictions, label='forecast')

plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

```python
mse_arima = mean_squared_error(test_data,test_predictions)
rmse_arima = np.sqrt(mse_arima)
mse_arima , rmse_arima
```

## SARIMA model

```python
from statsmodels.tsa.arima_model import ARIMA
import pmdarima as pm
```

```python
data4 = data.copy()
```

```python
train_data = data4.iloc[:32,0] 
test_data = data4.iloc[32:,0]
train_data
```

```python
test_data.shape
```

```python
# Seasonal - fit stepwise auto-ARIMA
smodel = pm.auto_arima(train_data, start_p=1, start_q=1,
                       test='adf',
                       max_p=3, max_q=3,max_d=3,
                       m=4,
                       start_P=0, seasonal=True,
                       d=None, D=2, trace=True,
                       error_action='ignore',
                       suppress_warnings=True, 
                       stepwise=True)

smodel.summary()
```

```python
# Forecast
fc,conf = smodel.predict(10, return_conf_int=True)
test_predictions = pd.Series(fc, index=test_data.index)
test_predictions
```

```python
# Plot
plt.figure(figsize=(12,5), dpi=100)
plt.plot(train_data, label='training')
plt.plot(test_data, label='test')
plt.plot(test_predictions, label='forecast')

plt.title('Forecast vs Actuals')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

```python
mse_sarima = mean_squared_error(test_data,test_predictions)
rmse_sarima = np.sqrt(mse_sarima)
mse_sarima , rmse_sarima
```
